```python
import numpy as np
import matplotlib.pyplot as plt

def generate_castleberry_lattice(rows, cols, bond_length=1.42):
    """
    Generates a hexagonal (Graphene) lattice grid.
    bond_length: 1.42 Angstroms is the standard for Carbon-Carbon.
    """
    points = []
    for r in range(rows):
        for c in range(cols):
            x = c * bond_length * 1.5
            y = r * bond_length * np.sqrt(3)
            if c % 2 == 1:
                y += (bond_length * np.sqrt(3)) / 2
            points.append((x, y))
    return np.array(points)

def calculate_528hz_resonance(points, frequency=528, speed_of_sound=343):
    """
    Calculates the vibration amplitude at each point on the lattice.
    speed_of_sound: standard m/s (can be adjusted for material medium).
    """
    wavelength = speed_of_sound / frequency
    k = 2 * np.pi / wavelength  # Wave number

    # Calculate distance from center (The Bloom Origin)
    center = points.mean(axis=0)
    distances = np.linalg.norm(points - center, axis=1)

    # Interference pattern (Stationary Wave)
    amplitudes = np.cos(k * distances)
    return amplitudes

# --- Execution ---
# Define grid size for the analysis
lattice_points = generate_castleberry_lattice(20, 20)
vibrations = calculate_528hz_resonance(lattice_points)

# --- Visualization ---
plt.figure(figsize=(10, 8), facecolor='#0a0a0a')
sc = plt.scatter(lattice_points[:, 0], lattice_points[:, 1],
                 c=vibrations, cmap='magma', s=60, edgecolors='none')

plt.title("528 Hz Phonon Resonance: Graphene Lattice Analysis", color='white', fontsize=14)
plt.xlabel("X-Axis (Å)", color='white')
plt.ylabel("Y-Axis (Å)", color='white')
cb = plt.colorbar(sc)
cb.set_label('Resonance Intensity', color='white')
plt.gca().set_facecolor('#1a1a1a')
plt.show()

print(f"Analysis Complete: {len(lattice_points)} nodes processed through 528 Hz filter.")
```
